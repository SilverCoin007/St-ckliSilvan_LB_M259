# 2.2 Bestimmen Sie für jedes Feld einige relevanten statistischen Informationen wie beispielsweise Median und Standardabweichung.

```python
# Import Datenbank
import pandas as pd
import os

bitcoin = pd.read_csv(os.path.join("Datenset_LB_M259.csv"))
```

```python
# Median:
print("Median low:", bitcoin["low"].median())
print("Median high:", bitcoin["high"].median())
print("Median open:", bitcoin["open"].median())
print("Median close:", bitcoin["close"].median())
print("Median volume:", bitcoin["volume"].median())
print("")

# Mittel:
print("Mittel low:", bitcoin["low"].mean())
print("Mittel high:", bitcoin["high"].mean())
print("Mittel open:", bitcoin["open"].mean())
print("Mittel close:", bitcoin["close"].mean())
print("Mittel volume:", bitcoin["volume"].mean())
print("")

# Modus:
print("Modus low:", bitcoin["low"].mode())
print("Modus high:", bitcoin["high"].mode())
print("Modus open:", bitcoin["open"].mode())
print("Modus close:", bitcoin["close"].mode())
print("Modus volume:", bitcoin["volume"].mode())
print("")
```
